```python
from numpy import equal
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/ubizareie/python-DA/main/StudentsPerformance.csv.xls')
#print (df.info())

#df['parental level of education'].value_counts().plot.pie()
#plt.pie (radius = 2, outpct= '%f%%')
plt.axis = ('equal')
print(df['parental level of education'].value_counts())
values=df['parental level of education'].value_counts()
plt.pie(values,labels = values.index, radius =1, autopct ='%0.0f%%', shadow = 'true',explode =(0.1,0,0,0,0,0),startangle =135, wedgeprops = {'edgecolor':'white'}, colors= ['#4287f5', '#f542e9', '#42f54e','#f59942','#42f59c','#f55142'])
plt.title ('Parental level of education', c ='blue', weight= 'extra bold')
plt.show()
```
